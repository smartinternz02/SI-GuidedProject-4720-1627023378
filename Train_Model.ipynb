```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Convolution2D,MaxPooling2D,Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from google.colab import drive 
drive.mount('/content/gdrive')
```

```python
train_1 = ImageDataGenerator(rescale = 1./255, shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test1 = ImageDataGenerator(rescale= 1./255)
```

```python
X_train = train_1.flow_from_directory(r'/content/gdrive/MyDrive/Smart_Bridge Project/data_set/train',target_size=(64,64),batch_size=128,class_mode="categorical")
X_test = test1.flow_from_directory(r'/content/gdrive/MyDrive/Smart_Bridge Project/data_set/val',target_size=(64,64),batch_size=128,class_mode="categorical")
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape=(64,64,3)))
```

```python
model.add(MaxPooling2D((2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(32,activation="relu"))



model.add(Dense(10,activation="softmax"))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train,steps_per_epoch = 18340/256 , epochs = 10 , validation_data = X_test,validation_steps = 7857/256)
```
